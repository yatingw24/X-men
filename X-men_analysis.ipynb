```python
import pandas as pd
import re
```

```python
df_xmen = pd.read_csv('x-men.csv')
df_xmen
```

### adding a new column called 'gender'

```python
df_xmen['gender'] = 'Unknown' 

df_xmen
```

```python
df_xmen['gender'] = ['Male', 'Female', 'Female', 'Male' , 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male']
```

```python
df_xmen['gender'] = df_xmen['Member'].apply(lambda x: 'Female' if x in [
    'jeanGrey', 'lornaDane', 'ororoMunroe', 'kittyPryde', 
    'annaMarieLeBeau', 'rachelSummers', 'alisonBlaire', 
    'jubilationLee', 'betsyBraddock'
] else 'Male')
```

```python
df_xmen
```

### narrow it down to only female x-men

```python
df_female = df_xmen[df_xmen['gender'] == 'Female']

df_female
```

### reformating each female x-man's name into a standard one

```python
def reformat(name):
    names = re.sub(r'([a-z])([A-Z])', r'\1 \2', name)
    return names.title() 

df_female['Member'] = df_female['Member'].apply(reformat)

df_female
```

### Analysis of Women's Market Value

```python
df_female[['Member', 'PPI80s_ebay', 'PPI90s_ebay', 'PPI80s_heritage', 'PPI90s_heritage']]
```

changing everything into a numeric value

```python
df_female['PPI80s_ebay'] = df_female['PPI80s_ebay'].astype(str).str.replace('[$,]', '', regex=True).astype(float)
df_female['PPI90s_ebay'] = df_female['PPI90s_ebay'].astype(str).str.replace('[$,]', '', regex=True).astype(float)


df_female['PPI80s_heritage'] = df_female['PPI80s_heritage'].astype(str).str.replace('[$,]', '', regex=True).astype(float)
df_female['PPI90s_heritage'] = df_female['PPI90s_heritage'].astype(str).str.replace('[$,]', '', regex=True).astype(float)
```

```python
df_female['drop_8090s_eBay'] = df_female['PPI90s_ebay'] / df_female['PPI80s_ebay']

df_female[['Member', 'PPI70s_ebay']]
```

```python
df_female['drop_8090s_heritage'] = df_female['PPI90s_heritage'] / df_female['PPI80s_heritage']

df_female[['Member', 'PPI80s_heritage', 'PPI90s_heritage','drop_8090s_heritage']]
```

```python
df_female[['Member', 'drop_8090s_eBay','90s_Appearance_Percent'] ]
```

```python
df_ebay_value = df_female[['Member', 'PPI80s_ebay','80s_Appearance_Percent']]
```

```python
df_ebay_value.to_csv('febay_output.csv', index = False)
```
